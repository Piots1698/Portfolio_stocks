```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read stock returns
ticker = "RIOT"
fcsv = f"https://raw.githubusercontent.com/multidis/hult-ai-investment-management/main/financial_data/{ticker}_daily_returns.csv"
sA = pd.read_csv(fcsv, index_col=[0])
sA
```

```python
ticker = "MARA"
fcsv = f"https://raw.githubusercontent.com/multidis/hult-ai-investment-management/main/financial_data/{ticker}_daily_returns.csv"
sB = pd.read_csv(fcsv, index_col=[0])
sB
```

```python
def selperiod(s):
    return s.loc[(s.index >= "2022-12-01") & (s.index <= "2022-12-31")]
```

```python
selperiod(sA)
```

```python
selperiod(sB)
```

```python
sA.plot(title = "RIOT")
```

```python
# average returns
sA.mean(), sB.mean()
```

```python
# volatilities
sA.std(), sB.std()
```

```python
# Portfolio returns: long A, short B (note: slightly oversimplified)
pAB = 0.5*selperiod(sA) - 0.5*selperiod(sB)
pAB
```

```python
# cumulative equity curve
pAB_equity = (1 + pAB).cumprod() - 1
pAB_equity
```

```python
pAB_equity.plot(title = "Long RIOT, Short MARA")
```

```python
pAB_equity.index = pd.to_datetime(pAB_equity.index)
```
